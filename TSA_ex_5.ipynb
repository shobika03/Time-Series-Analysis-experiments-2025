```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
# 1. Create sample time series data
np.random.seed(0)
n = 100
t = np.arange(n)
trend = 5 + 0.4 * t  # Linear trend
noise = np.random.normal(0, 2, n)
y = trend + noise
```

```python
# Create DataFrame
df = pd.DataFrame({'Time': t, 'Value': y})
```

```python
# 2. Prepare features (X) and labels (y)
X = df[['Time']]  # Independent variable (time)
y = df['Value']   # Dependent variable (value)
```

```python
# 3. Train-Test Split (80%-20%)
split = int(n * 0.8)
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# 4. Train Linear Regression Model
model = LinearRegression()
model.fit(X_train, y_train)

# 5. Predict
y_pred = model.predict(X_test)

# 6. Forecast future (e.g., next 10 time steps)
future_t = np.arange(n, n+10).reshape(-1, 1)
future_pred = model.predict(future_t)
```

```python
# 7. Plotting
plt.figure(figsize=(12, 6))
plt.plot(t, y, label='Original Data')
plt.plot(X_test['Time'], y_pred, color='red', label='Predicted Test Data')
plt.plot(future_t, future_pred, color='green', linestyle='dashed', label='Forecast Future')
plt.title('Linear Regression Forecasting')
plt.xlabel('Time')
plt.ylabel('Value')
plt.legend()
plt.grid(True)
plt.show()
```
