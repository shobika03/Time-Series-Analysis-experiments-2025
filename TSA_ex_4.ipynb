```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Create sample time series data
date_range = pd.date_range(start='2020-01-01', periods=100, freq='D')
trend = np.linspace(10, 50, 100)  # Linear trend
noise = np.random.normal(0, 3, 100)  # Random noise
data = trend + noise
```

```python
df = pd.DataFrame({'Date': date_range, 'Value': data})
df.set_index('Date', inplace=True)
```

```python
# 2. Aggregation: Resample data (e.g., to weekly average)
weekly_avg = df['Value'].resample('W').mean()
```

```python
df['Moving_Avg'] = df['Value'].rolling(window=7).mean()

# 4. Detrending: Subtracting the moving average
df['Detrended'] = df['Value'] - df['Moving_Avg']
```

```python
# 5. Plotting
plt.figure(figsize=(14, 8))

plt.subplot(3, 1, 1)
plt.plot(df['Value'], label='Original Data')
plt.title('Original Time Series')
plt.legend()

plt.subplot(3, 1, 2)
plt.plot(df['Moving_Avg'], color='orange', label='Smoothed (Moving Average)')
plt.title('Smoothed Trend')
plt.legend()

plt.subplot(3, 1, 3)
plt.plot(df['Detrended'], color='green', label='Detrended Series')
plt.title('Trend Eliminated (Detrended)')
plt.legend()

plt.tight_layout()
plt.show()
```
