```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.stattools import adfuller
```

```python
np.random.seed(42)
date_rng = pd.date_range(start='1/1/2020', periods=100, freq='D')
data = np.cumsum(np.random.randn(100))  # Non-stationary data (random walk)
```

```python
df = pd.DataFrame({'Date': date_rng, 'Value': data})
df.set_index('Date', inplace=True)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df, label="Time Series")
plt.title("Time Series Data")
plt.legend()
plt.show()
```

```python
def check_stationarity(timeseries):
    result = adfuller(timeseries)
    print("ADF Statistic:", result[0])
    print("p-value:", result[1])
    print("Critical Values:", result[4])

    if result[1] < 0.05:
        print("Conclusion: The data is stationary.")
    else:
        print("Conclusion: The data is non-stationary.")


check_stationarity(df['Value'])
```
