```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

df = pd.DataFrame({
    'Date': pd.date_range(start='2025-01-01', periods=50, freq='D'),
    'Value': range(50)
})
df.set_index('Date', inplace=True)
```

```python

plt.figure(figsize=(10, 5))
plt.plot(df.index, df['Value'], marker='o', linestyle='-', color='r', label='Value')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Line Plot')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(df.index, df['Value'], color='b', label='Value')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Scatter Plot')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.bar(df.index, df['Value'], color='violet', label='Value')
plt.xlabel('Date')
plt.ylabel('Value')
plt.title('Time Series Bar Plot')
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.show()
```

```python
from pandas.plotting import autocorrelation_plot

plt.figure(figsize=(10, 5))
autocorrelation_plot(df['Value'])
plt.title('Autocorrelation Plot')
plt.show()
```
